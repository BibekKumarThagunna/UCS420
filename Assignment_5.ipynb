Q1.**For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find**
1.**Sum of all elements**

2.**Sum of all elements row-wise**

3.**Sum of all elements column-wise**

```python
import numpy as np
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
print(gfg)
print(type(gfg))
print('Sum of all elements is:',np.sum(gfg))
print('Sum of all elements row wise is:',np.sum(gfg,axis=1).T)
print('Sum of all elements column wise is:',np.sum(gfg,axis=0))
```

**Q2. (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find**
          
i.**Sorted array**

ii.**Indices of sorted array**

iii.**4 smallest elements**

iv.**5 largest elements**

**(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find**

i.**Integer elements only**

ii.**Float elements only**

```python
array=np.array([10,52,62,16,16,54,453])
print('Sorted array is:',np.sort(array))
print('Indices of sorted array',np.argsort(array))
print('4 smallest elements are:',np.sort(array)[:4])
print('5 largest elements are:',np.sort(array)[-5:])

array=np.array([1.0,1.2,2.2,2.0,3.0,2.0])
print('Integer elements are:',array[array%1==0])
print('Float elements are:',array[array%1!=0])
```

**Q3.You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting.**
**Generate your unique sales dataset:**

**Take the sum of the ASCII values of the initials of your first and last name.Call this value X.**
**(If your initials are A B → ASCII sum =65+66=131→sales=[131,181,231,281,331].)**
**Create a NumPy array sales with values [X,X+50,X+100,X+150,X+200].**
**Compute your personalized tax rate as ((X % 5) + 5)/100.**

**Use broadcasting to apply this tax rate to each sales value.**
**Adjust sales based on discount:**

**If sales < X+100, apply a 5% discount.**
**If sales >= X+100, apply a 10% discount.**
**Expand sales data for multiple weeks:**

**Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting.**

**Increase sales by 2% per week using element-wise broadcasting.**

```python
x=ord('R')+ord('B')
sales=np.array([x,x+50,x+100,x+150,x+200])
print('Unique sales dataset is',sales)

tax_rate=((x%5)+5)/100
tax_on_sales=sales*(tax_rate)
print('Personalized tax rate is:',tax_rate)
print('Tax on sales:',tax_on_sales)

sales_after_discount=np.where(sales<x+100,sales*0.95,sales*0.90)
print('Sales after discount applied is:',sales_after_discount)

sales_with_tax=sales_after_discount*(1+tax_rate)
print('Sales after tax:',sales_with_tax)

weekly_sales=np.vstack([sales_with_tax]*3)
weeks=np.arange(3).reshape(-1,1)
adjusted_sales=weekly_sales*(1.02**weeks) 
print('3 weeks sales data is given below')
print(adjusted_sales)
```

**Q4.Generate x values using np.linspace() from -10 to 10 with 100 points. Use each function from the list below and compute y values using NumPy:**

**Y = x^2**

**Y = sin(x)**

**Y = e^x**

**Y = log(|x| + 1)**

**Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.**

```python
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
fig,p=plt.subplots(2,2,figsize=(10,6))
y=np.array(x*x)
p[0,0].plot(x,y,color='r')
p[0,0].set_title('$y=x^2$')
p[0,0].grid(True)
p[0,0].axhline(0,color='black',linewidth=2)
p[0,0].axvline(0,color='black',linewidth=2)
p[0,0].set_xlabel('x --->')
p[0,0].set_ylabel('y --->')
y=np.array(np.sin(x))
p[0,1].plot(x,y,color='b')
p[0,1].set_title('y=sin(x)')
p[0,1].grid(True)
p[0,1].axhline(0,color='black',linewidth=2)
p[0,1].axvline(0,color='black',linewidth=2)
p[0,1].set_xlabel('x --->')
p[0,1].set_ylabel('y --->')
y=np.array(np.exp(x))
p[1,0].plot(x,y,color='g')
p[1,0].set_title('$y=e^x$')
p[1,0].grid(True)
p[1,0].axhline(0,color='black',linewidth=2)
p[1,0].axvline(0,color='black',linewidth=2)
p[1,0].set_xlabel('x --->')
p[1,0].set_ylabel('y --->')
y=np.array(np.log(np.absolute(x)+1))
p[1,1].plot(x,y,color='y')
p[1,1].set_title('y=log(|x|+1)')
p[1,1].grid(True)
p[1,1].axhline(0,color='black',linewidth=2)
p[1,1].axvline(0,color='black',linewidth=2)
p[1,1].set_xlabel('x --->')
p[1,1].set_ylabel('y --->')
fig.tight_layout()
plt.show()
```
